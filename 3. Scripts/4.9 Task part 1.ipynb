```python
# Checking for mixed type data

for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

## Importing orders/products df for merge

```python
df_merged = pd.read_pickle(os.path.join(r'/content/drive/MyDrive/AB Instacart Project Folder/2. Data/Orders_products_merged_4_8.pkl'))
```

```python
# Checking datatypes

df_merged.dtypes
```

```python
# deleting unnecessary columns

df_merged_clean = df_merged.drop(columns = ['_merge','Unnamed: 0'])
```

```python
# Changing data types to minimize memory errors

df_merged_clean['order_id'] = df_merged_clean['order_id'].astype('int8')
df_merged_clean['first_order'] = df_merged_clean['first_order'].astype('int8')
df_merged_clean['product_id'] = df_merged_clean['product_id'].astype('int8')
df_merged_clean['add_to_cart_order'] = df_merged_clean['add_to_cart_order'].astype('int8')
df_merged_clean['aisle_id'] = df_merged_clean['aisle_id'].astype('int8')
df_merged_clean['department_id'] = df_merged_clean['department_id'].astype('int8')
df_merged_clean['prices'] = df_merged_clean['prices'].astype('float16')
df_merged_clean['average_order'] = df_merged_clean['average_order'].astype('float16')
```

```python
# Merging df's

df_full_merge = df_cust.merge(df_merged_clean, on = 'user_id')
```
